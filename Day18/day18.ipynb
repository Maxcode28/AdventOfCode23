```python
class Pit:
    def __init__(self, r, c) -> None:
        self.trench = "."
        self.colour = "#000000"
        self.row = r
        self.column = c
        self.crossover = False
        pass

    def dig(self, colour, shape):
        self.colour = colour
        if self.trench not in  [".", "L", "R"]:
            self.crossover = True
        self.trench = shape
        
    def inside(self):
        if self.trench == ".":
            self.trench = "#"
```

```python
instructions = []
lagoon = []
num_rows = 1
min_r = 0
max_r = 1
num_col = 1
min_c = 0
max_c = 1
with open(r"input-18.txt", "r") as file:
#with open(r"test-input-18.txt", "r") as file:
    for r_num, row in enumerate(file):
        row = row.replace("\n", "")
        row = row.replace("(","")
        row = row.replace(")", "")
        elements = row.split(" ")
        if elements[0] == "D":
            num_rows += int(elements[1])
            if num_rows > max_r:
                max_r = num_rows
        elif elements[0] == "U":
            num_rows -= int(elements[1])
            if num_rows < min_r:
                min_r = num_rows
        elif elements[0] == "R":
            num_col += int(elements[1])
            if num_col > max_c:
                max_c = num_col
        elif elements[0] == "L":
            num_col -= int(elements[1])
            if num_col < min_c:
                min_c = num_col
        instructions.append(elements)

num_rows = max_r - min_r + 1
num_col = max_c - min_c + 1

for r in range(num_rows):
    rg = []
    for c in range(num_col):
        rg.append(Pit(r,c))
    lagoon.append(rg)
current_r = -min_r + 1
current_c = -min_c + 1

for i in instructions:
    for steps in range(1,int(i[1]) + 1):
        if i[0] == "U" and current_r > 0:
            current_r -= 1 
        if i[0] == "D"and current_r < num_rows:
            current_r += 1
        if i[0] == "L" and current_c > 0:
            current_c -= 1
        if i[0] == "R" and current_c < num_col:
            current_c += 1
        
        lagoon[current_r][current_c].dig(i[2],i[0])
        if steps == int(i[1]) and i[0] in ["R", "L"]:
            lagoon[current_r][current_c].dig(i[2],"C")

for j,p in enumerate(lagoon):
    toggle = False
    toggles = 0
    last_symbol = ""
    for idx, i in enumerate(p):
        if i.trench == "C":
            if j > 0:
                if lagoon[j-1][idx].trench == "U":
                    i.trench = "U"
            if j < (max_r - min_r - 1):
                if lagoon[j+1][idx].trench == "D":
                    i.trench = "D"
        if (i.trench in ["U", "D", "C"]) and i.crossover == False and i.trench != last_symbol:
            last_symbol = i.trench
            toggle = not toggle
            toggles += 1
        if toggle:
            i.inside()
count = 0
for l in lagoon:
    vis = []
    for i in l:
        if i.trench != ".":
            count += 1
        vis.append(i.trench)
    #print(vis)
print("Part 1, area", count)
```
