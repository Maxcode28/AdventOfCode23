```python
from enum import Enum
```

```python
card_symbols = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
card_symbols_2 = ['J','2', '3', '4', '5', '6', '7', '8', '9', 'T', 'Q', 'K', 'A']
part2 = True

class GameType(Enum):
    FiveOfAKind = 7
    FourOfAKind = 6
    FullHouse = 5
    ThreeOfAKind = 4
    TwoPair = 3
    OnePair = 2
    HighCard = 1

class Card:
    def __init__(self, symbol):
        self.symbol = symbol
        self.strength = self._get_strength()

    def _get_strength(self):
        if not part2:
            for i,c in enumerate(card_symbols):
                if self.symbol == c:
                    return i
        else:
            for i,c in enumerate(card_symbols_2):
                if self.symbol == c:
                    return i            

class Hand:
    def __init__(self, hand, bet):
        self.hand = hand
        self.strength = []
        self.type = GameType
        self.cards = self._get_cards(hand)
        self.bet = int(bet)
        self._get_strengths()
        self.winnings = 0

    def _get_cards(self, hand):
        hand_l = list(hand)
        cards = []
        for i,c in enumerate(hand_l):
            cards.append(Card(c))
        if not part2:
            self._get_hand_type(hand_l)
        else:
            self._get_hand_type_2(hand_l)
        return cards
    
    def _get_hand_type(self, hand):
        hand = sorted(hand)
        prev_val = ""
        occ = 1
        most_occ = 0
        s_most_occ = 0
        for idx, h in enumerate(hand):
            if h == prev_val:
                occ = occ + 1
            if h != prev_val or idx == len(hand) - 1:
                if occ > most_occ:
                    s_most_occ = most_occ
                    most_occ = occ
                    occ = 1
                elif occ > s_most_occ:
                    s_most_occ = occ
                    occ = 1
                else:
                    occ = 1
            prev_val = h
        if most_occ == 5:
            self.type = GameType.FiveOfAKind
        elif most_occ == 4:
            self.type = GameType.FourOfAKind
        elif most_occ == 3 and s_most_occ == 2:
            self.type = GameType.FullHouse
        elif most_occ == 3:
            self.type = GameType.ThreeOfAKind
        elif most_occ == 2 and s_most_occ == 2:
            self.type = GameType.TwoPair
        elif most_occ == 2:
            self.type = GameType.OnePair
        else:
            self.type = GameType.HighCard
        self.strength.append(int(self.type.value))

    def _get_hand_type_2(self, hand):
        hand = sorted(hand)
        prev_val = ""
        occ = 1
        most_occ = 0
        s_most_occ = 0
        jokers = 0
        for idx, h in enumerate(hand):
            if h == "J":
                jokers = jokers + 1
            elif h == prev_val:
                occ = occ + 1
            if h != prev_val or (idx == (len(hand) - 1)) or h == "J":
                if occ > most_occ:
                    s_most_occ = most_occ
                    most_occ = occ
                    occ = 1
                elif occ > s_most_occ:
                    s_most_occ = occ
                    occ = 1
                else:
                    occ = 1
            prev_val = h
        most_occ = most_occ + jokers
        if jokers == 5:
            most_occ = 5
        if most_occ == 5:
            self.type = GameType.FiveOfAKind
        elif most_occ == 4:
            self.type = GameType.FourOfAKind
        elif most_occ == 3 and s_most_occ == 2:
            self.type = GameType.FullHouse
        elif most_occ == 3:
            self.type = GameType.ThreeOfAKind
        elif most_occ == 2 and s_most_occ == 2:
            self.type = GameType.TwoPair
        elif most_occ == 2:
            self.type = GameType.OnePair
        else:
            self.type = GameType.HighCard
        self.strength.append(int(self.type.value))

    def _get_strengths(self):
        for c in self.cards:
            self.strength.append(c.strength)

        
        
```

```python
hands = []
with open(r"input-7.txt", "r") as file:
#with open(r"test-input-7.txt", "r") as file:
    for r_num, row in enumerate(file):
        row = row.replace("\n", "")
        s_row = row.split(" ")
        hands.append(Hand(s_row[0], s_row[1]))

    hands.sort(key=lambda x: x.strength, reverse=False)
    total_winnings = 0
    for i,h in enumerate(hands,1):
        h.winning = i*h.bet
        total_winnings = total_winnings + h.winning
    print(total_winnings)
```
