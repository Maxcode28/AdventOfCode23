```python
class Seed:
    def __init__(self, seed_value):
        self.seed = seed_value
        self.soil = seed_value
        self.fert = seed_value
        self.water = seed_value
        self.light = seed_value
        self.temp = seed_value
        self.humid = seed_value
        self.loc = seed_value

        self.updated_values = [False] * 7

    def _update(self, source, s_start, d_start, rg):
        if (source < s_start) or (source > (s_start + rg - 1)):
            return [False, source]
        else:
            diff = source - s_start
            upd_val = d_start + diff
            return [True, upd_val]

    def update_mapping(self, map_d, d_start, s_start, rg):
        update_succeeded = False
        if map_d == "soil":
            update_succeeded = self._update_soil(s_start, d_start, rg)
        elif map_d == "fertilizer":
            update_succeeded = self._update_fert(s_start, d_start, rg)
        elif map_d == "water":
            update_succeeded = self._update_water(s_start, d_start, rg)
        elif map_d == "light":
            update_succeeded = self._update_light(s_start, d_start, rg)
        elif map_d == "temperature":
            update_succeeded = self._update_temp(s_start, d_start, rg)
        elif map_d == "humidity":
            update_succeeded = self._update_humid(s_start, d_start, rg)
        elif map_d == "location":
            update_succeeded = self._update_loc(s_start, d_start, rg)
        return update_succeeded


    def _update_soil(self, s_start, d_start, rg):
        upd_res = self._update(self.seed, s_start, d_start, rg)
        if upd_res[0]:
            self.soil = upd_res[1]
            self.updated_values[0] = True
            return True
        if not self.updated_values[0]:
            self.soil = self.seed
        return False
        
    def _update_fert(self, s_start, d_start, rg):
        upd_res = self._update(self.soil, s_start, d_start, rg)
        if upd_res[0]:
            self.fert = upd_res[1]
            self.updated_values[1] = True
            return True
        if not self.updated_values[1]:
            self.fert = self.soil
        return False

    def _update_water(self, s_start, d_start, rg):
        upd_res = self._update(self.fert, s_start, d_start, rg)
        if upd_res[0]:
            self.water = upd_res[1]
            self.updated_values[2] = True
            return True
        if not self.updated_values[2]:
            self.water = self.fert
        return False
    
    def _update_light(self, s_start, d_start, rg):
        upd_res = self._update(self.water, s_start, d_start, rg)
        if upd_res[0]:
            self.light = upd_res[1]
            self.updated_values[3] = True
            return True
        if not self.updated_values[3]:
            self.light = self.water
        return False
    
    def _update_temp(self, s_start, d_start, rg):
        upd_res = self._update(self.light, s_start, d_start, rg)
        if upd_res[0]:
            self.temp = upd_res[1]
            self.updated_values[4] = True
            return True
        if not self.updated_values[4]:
            self.temp =self.light
        return False
    
    def _update_humid(self, s_start, d_start, rg):
        upd_res = self._update(self.temp, s_start, d_start, rg)
        if upd_res[0]:
            self.humid = upd_res[1]
            self.updated_values[5] = True
            return True
        if not self.updated_values[5]:
            self.humid = self.temp
        return False
    
    def _update_loc(self, s_start, d_start, rg):
        upd_res = self._update(self.humid, s_start, d_start, rg)
        if upd_res[0]:
            self.loc = upd_res[1]
            self.updated_values[6] = True
            return True
        if not self.updated_values[6]:
            self.loc = self.humid
        return False
```

```python
# Part 2: reverse mapping more efficient than mapping seeds to locations 
class Mapping:
    def __init__(self, map_name):
        self.name = map_name
        self.map = []

    def input_map(self, d_start, s_start, ran):
        self.map.append([d_start, s_start, ran])
        

    def find_source_for_dest_in_map(self, val):
        for m in self.map:
            diff = val - m[0]
            if diff >= 0 and diff < m[2]:
                source = m[1] + diff
                return source
        return val

class SeedRange:
    def __init__(self,s,e):
        self.start = s
        self.end = e
    
    def contains_source(self, source):
        if source >= self.start and source <= self.end:
            return True
        return False


    
```
